```python
import pandas as pd
pd.set_option('display.max_column',30)
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pylab import rcParams ##matplotlib
rcParams['figure.figsize'] = 12, 8
```

```python
train = pd.read_csv('first_round_training_data.csv')[['Parameter'+str(i) for i in range(1,11)]+['Quality_label']]
# 读取初赛需要的特征
test = pd.read_csv('first_round_testing_data.csv')
```

```python
train['Quality_label'].value_counts().plot(kind='bar')
# pycharm显示jupyter notebook的图需要在pycharm上面看
plt.show()
# 要画出来需要加show
```

```python
train['label'] = train['Quality_label'].map({
    'Excellent':1,
    'Good':2,
    'Pass':3,
    'Fail':4
})
```

```python
train.describe()
# 第一四分位数 (Q1)，又称“较小四分位数”，等于该样本中所有数值由小到大排列后第25%的数字。

# 第二四分位数 (Q2)，又称“中位数”，等于该样本中所有数值由小到大排列后第50%的数字。

# 第三四分位数 (Q3)，又称“较大四分位数”，等于该样本中所有数值由小到大排列后第75%的数字。
```

```python
plt.figure(figsize=(12,8))
correlation = train.corr()
mask = np.zeros_like(correlation)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data=correlation.round(3),mask=mask,vmin=-0.1,vmax=0.1,annot=True, linewidths=.1)
plt.show()
# 看变量之间的相关性
```

```python
train.nunique().plot(kind='bar')
# 判断是否是类别变量
plt.show()
# 如果类别多的就不是，类别少的就可能是类别变量
```
